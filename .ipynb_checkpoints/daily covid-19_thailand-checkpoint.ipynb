### read data from .xlsx

```python
# url='https://data.go.th/dataset/8a956917-436d-4afd-a2d4-59e4dd8e906e/resource/93f74e67-6f76-4b25-8f5d-b485083100b6/download/covid-19-daily.xlsx'
url='https://data.go.th/dataset/8a956917-436d-4afd-a2d4-59e4dd8e906e/resource/93f74e67-6f76-4b25-8f5d-b485083100b6/download/20200414_pm.xlsx'
# df = pd.read_excel(url)
df = pd.read_excel(url, na_values=['ไม่ระบุ', '-', 'ไม่ทราบ'],   
                   true_values=['YES'], false_values=['NO'],  
                    parse_dates=['Notification date', 'Announce Date'])
df
```
